```python
# imprime los tipos de datos de mi df; dara pistas sobre que tipo de algoritmo de ML usar
df.dtypes
# preg (Pregnancies) - numero de embarazos que ha tenido la paciente.
# plas (Plasma Glucose)- concentracion de glucosa en plasma, utilizada para evaluar niveles de azucar en sangre.
# pres (Blood Pressure)- presión arterial diastólica medida en mm Hg.
# skin (Skin Thickness)- grosor del pliegue cutáneo del tríceps en milímetros, relacionado con la grasa corporal.
# test (Insulin)- nivel de insulina sérica en sangre.
# mass (BMI - Body Mass Index)- índice de masa corporal, calculado a partir del peso y la altura.
# pedi (Diabetes Pedigree Function)- indicador que refleja la probabilidad de diabetes según antecedentes familiares.
# age (Age)- edad de la paciente en años.
# class - indica si la paciente tiene diabetes (1) o no (0).
```

```python
# asimetria -> indica si los datos estan cargados hacia un lado de la distribucion
df.skew()

# Entre -0.5 y 0.5	aproximadamente simetrica
# Entre -1 y -0.5 o 0.5 y 1	moderadamente asimetrica
# Menor que -1 o mayor que 1	fuertemente asimetrica
```
