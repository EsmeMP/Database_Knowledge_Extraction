```python
# asimetria -> indica si los datos estan cargados hacia un lado de la distribucion
df.select_dtypes(include='number').skew()

# Entre -0.5 y 0.5	aproximadamente simetrica
# Entre -1 y -0.5 o 0.5 y 1	moderadamente asimetrica
# Menor que -1 o mayor que 1	fuertemente asimetrica

# eye_diameter_mm (1.31) presenta una asimetría positiva, la mayoria de los peces tienen diametros de ojo relativamente pequeños, pero existen algunos ejemplares con ojos considerablemente mas grandes.
# total_length_cm (2.14), body_width_cm (2.59) y head_length_cm (2.33) muestran una asimetría positiva fuerte, esto indica que la mayor parte de los peces tiene tamaños moderados, mientras que unos pocos ejemplares son mucho más grandes que el resto.
```
