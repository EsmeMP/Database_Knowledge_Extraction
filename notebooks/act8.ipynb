# Limpieza de Datos - Dataset Titanic
Practicamos las técnicas principales: valores nulos, duplicados y conversión de tipos.

```python
# Importar librerías y cargar el dataset
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.head(10)
```

## Identificar valores nulos

```python
# Ver cuántos valores nulos hay en cada columna
print(df.isnull().sum())
```

```python
# Ver el porcentaje de nulos por columna
porcentaje_nulos = (df.isnull().sum() / len(df)) * 100
print(porcentaje_nulos.sort_values(ascending=False))
```

## Eliminar registros con valores faltantes

```python
# Eliminar TODAS las filas que tengan al menos un nulo
df_sin_nulos = df.dropna()

print(f'Filas antes:  {len(df)}')
print(f'Filas después: {len(df_sin_nulos)}')
print(f'Filas eliminadas: {len(df) - len(df_sin_nulos)}')
```

## Sustituir por la media — columna `age`

```python
# Ver nulos en 'age' antes
print('Nulos en age ANTES:', df['age'].isnull().sum())
print('Media de edad:', df['age'].mean().round(2))
```

```python
# Sustituir nulos de 'age' con la media
df['age'] = df['age'].fillna(df['age'].mean())

print('Nulos en age DESPUÉS:', df['age'].isnull().sum())
df['age'].describe()
```

## Sustituir por la mediana — columna `fare`

```python
# Ver nulos en 'fare' antes
print('Nulos en fare ANTES:', df['fare'].isnull().sum())
print('Mediana de fare:', df['fare'].median())
```

```python
# Sustituir nulos de 'fare' con la mediana
df['fare'] = df['fare'].fillna(df['fare'].median())

print('Nulos en fare DESPUÉS:', df['fare'].isnull().sum())
df['fare'].describe()
```

## Eliminación de duplicados

```python
# Ver cuántos duplicados hay
print('Filas duplicadas:', df.duplicated().sum())
```

```python
# Eliminar duplicados
df = df.drop_duplicates()

print('Filas después de drop_duplicates():', len(df))
print('Duplicados restantes:', df.duplicated().sum())
```

## Conversión de tipo de datos

```python
# Ver tipos de datos actuales
print(df.dtypes)
```

```python
# Convertir 'survived' de int a bool (tiene más sentido lógico)
print('Antes:', df['survived'].dtype)
df['survived'] = df['survived'].astype(bool)
print('Después:', df['survived'].dtype)
df['survived'].value_counts()
```

```python
# Convertir 'pclass' de int a string (es categoría, no número)
print('Antes:', df['pclass'].dtype)
df['pclass'] = df['pclass'].astype(str)
print('Después:', df['pclass'].dtype)
df['pclass'].value_counts()
```

## Resumen final del dataset limpio

```python
print('=== RESUMEN FINAL ===')
print(f'Filas: {len(df)}')
print(f'Columnas: {len(df.columns)}')
print(f'\nNulos por columna:')
print(df.isnull().sum())
print(f'\nTipos de datos:')
print(df.dtypes)
```
